dot product

```python
a=np.array([2,3])
b=np.array([4,4])
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

cross product

```python
a=np.array([2,3,6])
b=np.array([4,4,4])
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

projb a = (a ⋅ b) / ||b||² b

```python
a=np.array([2,3,6])
b=np.array([4,4,4])
magnitude_of_b=np.sqrt(sum(b**2))
a_projections_onto_b=np.dot(a,b)/(magnitude_of_b**2)*b
print(a_projections_onto_b)
```
